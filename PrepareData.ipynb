```python
import os
import random
import shutil
import numpy as np
```

```python
import zipfile

path_to_zip = 'archive.zip'
path_to_extract = 'extracted_data'
path_to_parsed_dataset = 'parsed_dataset'

train_perc = 0.8
validation_perc = 0.1
test_perc = 0.1
```

```python
if not os.path.exists(path_to_extract):
    os.makedirs(path_to_extract)

if not os.path.exists(path_to_parsed_dataset):
    os.makedirs(path_to_parsed_dataset)
```

```python
with zipfile.ZipFile(path_to_zip, 'r') as zip_ref:
    zip_ref.extractall(path_to_extract)
```

```python
folders = list(os.walk(path_to_extract))
```

```python

all_train_files = []
all_validation_files = []
all_test_files = []


for folder in folders:
    
    path_to_folder = folder[0]
    list_of_files = folder[2]
    
    if len(list_of_files) > 0:
        parking, weather, date = path_to_folder.split('\\')[-3:]
        relevant_files = []
        
        for file in list_of_files:
            file_name, file_type = file.split('.')
            
            if file_type == 'jpg':
                relevant_files.append(os.path.join(path_to_folder, file_name))
        
        amount_train = int(len(relevant_files) * train_perc)
        amount_validation = int(len(relevant_files) * validation_perc)
        amount_test = int(len(relevant_files) * test_perc)
        
        random.shuffle(relevant_files)
        
        train_files = relevant_files[:amount_train]
        validation_files = relevant_files[amount_train: amount_train + amount_validation]
        test_files = relevant_files[amount_train + amount_validation: ]
        
        print('train_files: ')
        print('\t', train_files)
        
        print('validation_files: ')
        print('\t', validation_files)
        
        print('test_files: ')
        print('\t', test_files)
        
        all_train_files += train_files
        all_validation_files += validation_files
        all_test_files += test_files
```

```python
len(all_train_files), len(all_validation_files), len(all_test_files)
```

```python
def copy_files_to_folder(path, file_list):
    if not os.path.exists(path):
        os.makedirs(path)
    
    for file in file_list:
        try:
            shutil.copyfile(file + '.jpg', os.path.join(path, file.split('\\')[-1]) +'.jpg')
            shutil.copyfile(file + '.xml', os.path.join(path, file.split('\\')[-1]) +'.xml')
        except Exception as e:
            print(e)
```

```python
copy_files_to_folder( os.path.join(path_to_parsed_dataset, 'train'),  all_train_files)
copy_files_to_folder( os.path.join(path_to_parsed_dataset, 'validation'), all_validation_files )
copy_files_to_folder( os.path.join(path_to_parsed_dataset, 'test'), all_test_files )
```
